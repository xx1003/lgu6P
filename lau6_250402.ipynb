```python
# tips의 평균 구하기
# 평균보다 큰 데이터만 조회

# pandas 내부에서 수치 연산은 전부 numpy 기반. 그래서 mean_tip이 type이 np
mean_tip = tips['tip'].mean()
mean_tip
# numpy와 차이점은 안에서 컬럼명 지정한 것 밖에 x
tips[tips['tip'] > mean_tip]
```

# loc vs iloc
- 코드 비교

## loc

```python
# tips.loc[tips['day'] == 'Sat']
# tips.loc[행, 열]
tips.loc[0:1, ['total_bill', 'tip', 'day']]
```

```python
tips.iloc[0:1, [0, 1, 4]]
```

```python
# tips.loc[행 조건식, 가져올 열]

tips.loc[tips['total_bill'] <= 11, :].reset_index(drop=True).head()
```

```python
tips.loc[tips['time'] == 'Dinner', :].reset_index(drop=True).head()
```

```python
tips.loc[tips['day'] == 'Sat', ['total_bill', 'tip', 'day']].reset_index(drop=True).head(1)
```

```python
tips.loc[(tips['time'] == 'Dinner') & (tips['total_bill'] <= 11), :].reset_index(drop=True)
# 조건식 2개
# tips.loc[(조건식1) & (조건식2), :]
```

```python
# iris
# 품종이 virginica 이거나 sepal_length >= 5인 값만 가져오고
# 컬럼은 sepal_length, petal_length, species만 가져오기
iris
iris.loc[ (iris['species'] == 'virginica') | (iris['sepal_length'] >= 5), ['sepal_length', 'petal_length', 'species']].reset_index(drop=True)
```
